```python
import requests
from bs4 import BeautifulSoup

base_url = "https://quotes.toscrape.com"
url = base_url

while True:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "lxml")

    for q in soup.select("div.quote"):
        tags = [t.get_text() for t in q.select("a.tag")]
        if "life" in tags:
            print("quote: " + q.select("span.text")[0].get_text())
            print("Author: " + q.select("small.author")[0].get_text())

    
    next_btn = soup.select_one("li.next a")
    print("")
    
    if next_btn:
        pg_num = next_btn["href"].split("/")
        print(pg_num[1]+pg_num[2])
        url = base_url + next_btn["href"]
    else:
        print("End of the Page")
        break
```
