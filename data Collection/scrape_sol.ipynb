```python
import requests
from bs4 import BeautifulSoup
```

```python
# fetch and Store
pg_cnt = 1

while True:
    url = f"https://quotes.toscrape.com/page/{pg_cnt}/"
    res = requests.get(url)

    soup = BeautifulSoup(res.text,"lxml")
    quotes = soup.select("div.quote")
    
    if not quotes:
        print("End")
        break

    with open(f"scrapped_data/quotes{pg_cnt}.html","w",encoding="utf-8") as f:
        f.write(res.text)
        print(f"downloaded data from page{pg_cnt}")
        
    pg_cnt = pg_cnt+1;
```

```python
# extract useful info
with open("scrapped_data/quotes1.html","r",encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html,"lxml")
```

```python
all_quotes = soup.select("div.quote")
life_quotes = []
for q in all_quotes:
    all_tags = []
    # print(q)
    for t in q.select("div.tags .tag"):
        # print(t)
        all_tags.append(t.get_text())
    
    if "life" in all_tags:
        text = q.select_one("span.text").get_text()
        author = q.select_one("small.author").get_text()
        life_quotes.append([text,author])

print(life_quotes)
```

```python
import os

life_quotes = []
cnt = 1
while True:
    file_path = f"scrapped_data/quotes{cnt}.html"
    if not os.path.exists(file_path):
        break 
        
    with open(file_path, "r", encoding="utf-8") as f:
        html = f.read()

        soup = BeautifulSoup(html,"lxml")
        all_quotes = soup.select("div.quote")
        
        for q in all_quotes:
            all_tags = []
            for t in q.select("div.tags .tag"):
                all_tags.append(t.get_text())
            
            if "life" in all_tags:
                text = q.select_one("span.text").get_text()
                author = q.select_one("small.author").get_text()
                life_quotes.append([text,author])
    cnt += 1

for q in life_quotes:
    print(q)
```
