```python
from bs4 import BeautifulSoup
```

```python
with open("scrapped_data/data1.html","r") as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content,"lxml")
# soup
```

```python
all_h3 = soup.findAll("h3")
all_h3
for h3 in all_h3:
    name = h3.get_text(strip = True)
    print("Country: " + name)
    # print(h3.find_parent("div").select("span.country-capital")) #returns list
    print("Capital: " + h3.find_parent("div").select("span.country-capital")[0].get_text(strip=True))
    print("Population: " + h3.find_parent("div").select("span.country-population")[0].get_text(strip=True))
    print("Area: " + h3.find_parent("div").select("span.country-area")[0].get_text(strip=True))
    print("")
    # print(name)
    # print(h3.find_parent("div")["class"])
    # print(h3.find_next("div").select("span.country-population")[0].get_text(strip=True))
```
