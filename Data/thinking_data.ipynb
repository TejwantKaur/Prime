```python
def clean_data(data):
    text_num = {"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        # clean data convert (one->1)
        # print(user['rating'])
        raw_rating = user['rating'].strip().lower() # strip remove white spaces " 5" -> "5"
        if(raw_rating in text_num):
            raw_rating = text_num[raw_rating]

        # print(raw_rating)
        user['rating'] = str(raw_rating) # stores integer value

        # handle missing data
        raw_age = user.get("age")
        if(raw_age == None):
            user['age'] = None

        # deduplication remove
        # if(user['name'].strip() in unique_users):
        #     continue    # if already exists
        # unique_users.add(user["name"])
        # cleaned_data.append(user)
        if(user['name'].strip() not in unique_users):
            unique_users.add(user["name"])
            cleaned_data.append(user)
        
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
def get_insights(data):
    # avg rating
    avg_rating = 0
    for user in data:
        avg_rating += float(user["rating"])
    print(f"avg rating: {avg_rating/len(data)}")

    # percentage of poor rating
    poor_rat = 0;
    for user in data:
        if(float(user['rating']) < 3):
            poor_rat += 1

    print(f"poor rating: {poor_rat/len(data)*100}%")
```

```python
get_insights(data)
```

```python
def get_recom(data):
    recommendations = []
    for user in data:
        curr_recom = {}
        curr_recom['name'] = user['name']

        if (float(user['rating']) >= 4):
            curr_recom['brand'] = "Apple"
        else:
            curr_recom['brand'] = "Samsung"
        recommendations.append(curr_recom)
        
    return recommendations
```

```python
get_recom(data)
```
